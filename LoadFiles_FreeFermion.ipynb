```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

## Spinful (sz = 0)
## 3D 
file_path = "./datas-spin/freefermion_pretraining/n_14_7_7_dim_3_Theta_0.50_Emax_40_twist_0.250_0.250_0.250" \
            +"/nlayers_2_modelsize_16_nheads_4_nhidden_32_damping_0.00100_maxnorm_0.00100_batch_8192/data.txt"
            
#file_path = "./datas-spin/freefermion_pretraining/n_38_19_19_dim_3_Theta_0.50_Emax_60_twist_0.250_0.250_0.250" \
#            +"/nlayers_2_modelsize_16_nheads_4_nhidden_32_damping_0.00100_maxnorm_0.00100_batch_8192/data.txt"
```

```python
with open(file_path, 'r') as file:
    lines = file.readlines()
data = [line.strip().split() for line in lines]
data = np.array(data, dtype=np.float64)
print(data.shape)
```

```python
for j in range(data.shape[0]):
    i, F, F_std, E, E_std, S, S_std = data[j]
    print("iter: %04d" % i,
            " F: %.6f (%.6f)" % (F, F_std),
            " E: %.6f (%.6f)" % (E, E_std),
            " S: %.6f (%.6f)" % (S, S_std),)
```

```python
def wrap_text(text, width):
    return '\n'.join([text[i:i+width] for i in range(0, len(text), width)])

#title_str = file_path.replace("/", "\n")
```

```python
fig = plt.figure(figsize=(8, 6))
title_str = wrap_text(file_path, 50)
plt.title(title_str)    
## F E S
plt.plot(data[:, 0], data[:, 1], marker='.')
plt.plot(data[:, 0], data[:, 3], marker='.')
plt.plot(data[:, 0], data[:, 5], marker='.')
plt.xlabel('epochs')
plt.ylabel('values')
plt.legend(['F','E','S'])
plt.show()
```
