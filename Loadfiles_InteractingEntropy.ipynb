```python
list_data = []

for file_path in [file_path1, file_path2, file_path3, file_path4, file_path5, file_path6]:
    with open(file_path, 'r') as file:
        lines = file.readlines()
    data = [line.strip().split() for line in lines]
    data = np.array(data, dtype=np.float64)
    print(data.shape)
    list_data.append(data)
```

```python
fig = plt.figure(figsize=(8, 6))
#title_str = wrap_text(file_path, 50)
plt.title('Entropy')    
## F E K V S
plt.plot([1, 3000],[1.643380,1.643380])
for i in range(len(list_data)):
    plt.plot(list_data[i][:, 0], list_data[i][:, 9])
plt.xlabel('epochs')
plt.ylabel('S')
plt.legend(['Free Fermion','rs=0.5','rs=0.8','rs=1.0','rs=2.0','rs=10.0','rs=40.0'])
plt.show()
```
